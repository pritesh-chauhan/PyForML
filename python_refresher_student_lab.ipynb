### Task 1.1: Comprehensions (no unnecessary loops)

Create (without loops):

- `pairs`: list of (x, x*x) for x=1..10
- `flattened`: flatten `nested`

**Explain:** Why can comprehensions be faster than manual loops?

```python
# TODO:
# Create even_squares
# HINT: range(1, 21), filter even with x % 2 == 0
# squares of even numbers from 1..20
even_squares = [x * x for x in range(1, 21) if x % 2 == 0]

# list of (x, x*x) for x = 1..10
pairs = [(x, x * x) for x in range(1, 11)]

# Flatten nested
nested = [[1, 2], [3, 4], [5, 6]]

flattened = [item for sublist in nested for item in sublist]

check('even_squares', even_squares == [4, 16, 36, 64, 100, 144, 196, 256, 324, 400])
check('pairs', pairs[:3] == [(1, 1), (2, 4), (3, 9)] and pairs[-1] == (10, 100))
check('flattened', flattened == [1, 2, 3, 4, 5, 6])

# TC: O(n) elements across sublists
# SC: O(n) number of elements in list

# list comprehensions push the looping logic down into optimized C code inside CPython while a manual for loop lives more in Python land
```

```python
from collections import Counter
s = 'mississippi'

# TODO:
freq = Counter(s)
top2 = freq.most_common(2)

print(freq)
print(top2)
check('top2', top2[0][0] == 'i' and top2[0][1] == 4)

# TC: O(n) considering single pass to count + linear-time selection for top 2
# SC: O(m) for unique chars

# Counter can be utilized when we need frequencies, or when we only need top-k elements. Sorting is about relative order.
```

### Task 2.1: Two Sum (Hash Map Pattern)

**FAANG Classic.** Return indices of two numbers that sum to target.

# HINT:
- Use a dict: value -> index
- For each `x`, check if `target-x` has been seen

**Explain:** Why is this O(n) time?

```python
def two_sum(nums, target):
    seen = {}
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

check('two_sum_1', two_sum([2, 7, 11, 15], 9) == [0, 1])
check('two_sum_2', two_sum([3, 2, 4], 6) == [1, 2])
check('two_sum_dup', two_sum([3, 3], 6) == [0, 1])

# TC: O(n) iterate all numbers
# SC: O(n)

# Since, we iterate the all the nums
```

### Task 2.2: Longest Substring Without Repeating Characters (Sliding Window)

**FAANG Classic/Gotcha:** naive O(n²) solutions time out.

# HINT:
- Keep a map char -> most recent index
- Maintain a window `[l..r]` with unique chars

**Explain:** What invariant does the window maintain?

```python
def length_of_longest_substring(s: str) -> int:
    if not s:
        return 0

    char_set = set()
    max_length = 0
    left = 0

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)

    return max_length


check('ls_abcabcbb', length_of_longest_substring('abcabcbb') == 3)
check('ls_bbbbb', length_of_longest_substring('bbbbb') == 1)
check('ls_pwwkew', length_of_longest_substring('pwwkew') == 3)
check('ls_empty', length_of_longest_substring('') == 0)

# The window always represents a substring with no repeated characters, and all updates to left and right are done to preserve that property.
```

### Task 2.3: Product of Array Except Self (Prefix/Suffix)

**FAANG Classic/Gotcha:**
- No division
- O(n) time

# HINT:
- Build prefix products in one pass
- Multiply by suffix products in reverse pass

**Explain:** Why does this work with zeros?

```python
def product_except_self(nums):
    # TODO
    ...
    # 1  1  2 6
    # 24 12 4 1
    # 24 12 8 6

    n = len(nums)
    prefix = [1] * n
    suffix = [1] * n
    result = [1] * n

    prefix[0] = 1
    for i in range(1, n):
        prefix[i] = prefix[i - 1] * nums[i - 1]
    suffix[n - 1] = 1
    for i in range(n - 2, -1, -1):
        suffix[i] = suffix[i + 1] * nums[i + 1]
    print(prefix)
    print(suffix)

    for i in range(n):
      result[i] = prefix[i] * suffix[i]
    return result

check('pes_1234', product_except_self([1, 2, 3, 4]) == [24, 12, 8, 6])
check('pes_zero', product_except_self([-1, 1, 0, -3, 3]) == [0, 0, 9, 0, 0])

#TC: O(n)
#SC: O(n)

# It works with zeros because the solution multiplies prefix and suffix products without division, and any product that includes a zero becomes zero. With one zero, only the zero’s index avoids multiplying by it; with multiple zeros, all results are zero automatically.
```

### Task 3.1: Sliding Window Maximum (Deque)

**FAANG Classic:** maintain a decreasing deque of indices.

# HINT:
- Pop from left if out of window
- Pop from right while current value is larger

**Explain:** Why is this amortized O(n)?

```python
def max_sliding_window(nums, k):
    # TODO
    if not nums or k <= 0:
        return []
    if k == 1:
        return nums[:]

    dq = deque()
    res = []

    for i, x in enumerate(nums):
        while dq and dq[0] <= i - k:
            dq.popleft()
        while dq and nums[dq[-1]] <= x:
            dq.pop()

        dq.append(i)
        if i >= k - 1:
            res.append(nums[dq[0]])

    return res

check('msw', max_sliding_window([1,3,-1,-3,5,3,6,7], 3) == [3,3,5,5,6,7])

# TC: O(n) each index is dequ'd once
# SC: O(k) holds at most k indices

# Because right only moves forward from 0→n once, and left also only moves forward from 0→n once. Even though the inner while can run many times, across the whole run each character is added to the set once and removed once, so total work is linear: amortized O(n).
```
